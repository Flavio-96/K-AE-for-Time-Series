```

from dtaidistance import dtw
import numpy as np
from google.colab import files
uploaded = files.upload()
```

```
import pandas as pd
import io
trainDf = pd.read_csv(io.BytesIO(uploaded['ECG5000_TRAIN.tsv']),sep='\t', header=None)
trainDf.head()
```

```
import struct
maxval = 2 ** (struct.Struct('i').size * 8 - 1) - 1
print(maxval)
```

```
from numpy import inf
ds = dtw.distance_matrix_fast(trainDf.to_numpy(), max_length_diff=maxval)
ds[ds == inf] = maxval
print(ds)
```

```
from sklearn.cluster import DBSCAN
from sklearn.preprocessing import StandardScaler
model = DBSCAN(metric="precomputed")
#X = StandardScaler().fit_transform(ds)
labels = model.fit_predict(ds)
labels
```
